```python
# standardize the features
scaler = StandardScaler()
df_std = scaler.fit_transform(df)


#let us do the same, but now reduce to 2 components
#The choice to use PCA(n_components=0.90) 
# means that you wanted to retain components that together explain at least 90% of the variance.
# However, the two components only explain 46.93% of the variance. 
# To reach 90%, you would need to include more components (as shown by the cumulative explained variance ratio).
# perform PCA
pca = PCA(n_components=0.90) 
df_pca = pca.fit_transform(df_std)
#   表示你希望保留的主成分能够解释原始数据方差的90%。
#   这意味着PCA会选择足够数量的主成分，使这些主成分的累积解释方差比达到或超过90%。#
# 或者 PCA(n_components=3) 会保留前三个主成分。

# the transformed data is an array, convert back to dataframe
df_pca = pd.DataFrame(df_pca)

#print the explained variance ratio
print('Explained variance ratio: ' , pca.explained_variance_ratio_)

cumsum_variance = np.cumsum(pca.explained_variance_ratio_)
print('cumulative explained variance ratio: ', cumsum_variance)

df_pca.head()



#
```

```python
#datatype conversion
bikes['datetime'] = pd.to_datetime(bikes['datetime'])

#extract hour, month, day into new columns
bikes['hour'] = bikes['datetime'].dt.hour
bikes['month'] = bikes['datetime'].dt.month
bikes['day'] = bikes['datetime'].dt.day
bikes.columns
```

```python
bikes['season'].unique()
```

```python
# create a new column with the bin of counts
# step 1: create a function to classify amounts, 
def total_cat(x):
    if x>= 0 and x<10:
        return '0-10'
    elif x>=10 and x<50:
        return '10-50'
    elif x>=50 and x<100:
        return '50-100'
    else:
        return '100+'
    
# step 2: create a new column 
# step 3: the .apply() method is used to apply the total_cat() function to each value in the 'count' column of the bikes DataFrame.
bikes['rental_total_group'] =bikes['count'].apply(total_cat)
```

```python
# create function with 2 inputs -temp and humidity -to classify goood/bad days
def good_bad(temp, hum):
    if temp > 25 and hum >70:
        return 'too hot'
    elif temp<=25 and hum <=70 and hum > 50:
        return 'so so day'
    else:
        return 'good day'

# apply function and create a new column 
bikes['day_type'] = bikes.apply(lambda x: good_bad(x['temp'], x['humidity']), axis =1)

#The .apply() method is used to apply a function along an axis (either rows or columns) of the DataFrame.

# another way to apply function
def application_function(x):
    good_bad(x['temp'], x['humidity'])
```
