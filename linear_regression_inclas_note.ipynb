```python
bikes.set_index('datetime').resample('h').agg({'humidity': 'mean', 'temp': 'mean', 'casual': 'sum'}) # group the data first
```

```python
bikes_resampled = bikes.set_index('datetime').resample('h').ffill()
```

```python
result_bikes_resampled = ols('casual~temp+humidity', data = bikes_resampled).fit()
print(result_bikes_resampled.summary())
```
